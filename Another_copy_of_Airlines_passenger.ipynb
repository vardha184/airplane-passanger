```python
def main3 (x,i):

  q1=np.percentile(x[i],25)
  median=np.percentile(x[i],50)
  q3=np.percentile(x[i],75)
  iqr=q3-q1
  lower_fence=q1-1.5*iqr
  upper_fence=q3+1.5*iqr
  x[i]=np.where(((x[i]<lower_fence)|(x[i]>upper_fence)),median,x[i])
for j in x.select_dtypes(include=['int','float']).columns:
  main3(x,j)
```
